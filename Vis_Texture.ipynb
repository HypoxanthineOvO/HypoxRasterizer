```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
IMG_PATH = "./assets/Money/2.png"

image = plt.imread(IMG_PATH)

# Display the image
plt.imshow(image)
plt.show()
```

```python
#  uv_res  UV  numpy 
# Read uv_res file
with open("uv_res", "r") as f:
    uv_res = np.array([list(map(float, line.strip().split())) for line in f.readlines()])

# visualize the UV coordinates
plt.imshow(image)
plt.scatter(uv_res[:, 0], uv_res[:, 1])
plt.show()
```
